```ruby
require 'active_support/all'
require 'nyaplot'
```

```ruby
numbers = (1..10000).map { rand }
```

```ruby
groups = numbers.group_by { |i| (i * 10).to_i }.map { |k, v| [k, v.size] }
```

```ruby
IRuby.table groups
```

```ruby
plot = Nyaplot::Plot.new
bar = plot.add(:bar, groups.map(&:first).map(&:to_s), groups.map(&:last))
plot.show
nil
```

```ruby
plot = Nyaplot::Plot.new
bar = plot.add(:bar, groups.map(&:first).map(&:to_s), groups.map(&:last))
bar.color(Nyaplot::Colors.qual)
plot.show
nil
```

```ruby
df = Nyaplot::DataFrame.new(group: groups.map(&:first).map(&:to_s), n: groups.map(&:last))
```

```ruby
plot = Nyaplot::Plot.new
plot.x_label("group")
plot.y_label("n")
bar = plot.add_with_df(df, :bar, :group, :n)
plot.show
nil
```

```ruby
df2 = Nyaplot::DataFrame.new(
  x: groups.map(&:first).map(&:to_s),
  y: groups.map(&:last),
  name: groups.map(&:last).map { |n| "#{n.to_f / 100}%"}
)
```

```ruby
groups2 = numbers.group_by { |i| ((i*30)**i).to_i }
groups2_sizes = groups2.map { |k, v| [k, v.size] }
```

```ruby
groups2_sums = groups2.map(&:last).map { |numbers| numbers.sum { |s| s.to_s.count('0') } / numbers.size.to_f }
```

```ruby
indexes = groups2_sizes.map(&:first).sort.map(&:to_s)
```

```ruby
plot = Nyaplot::Plot.new
bar = plot.add(:bar, indexes, groups2_sums)
plot.show
nil
```

```ruby
groups2_all_sums = ('0'..'9').map do |c|
  groups2.map(&:last).map { |numbers| numbers.sum { |s| s.to_s.count(c) } / numbers.size.to_f }
end
```

```ruby
plot = Nyaplot::Plot.new
bar = plot.add(:bar, indexes, groups2_sizes.map(&:last).map { |i| i.to_f / 100 })
10.times do |i|
  line = plot.add(:line, indexes, groups2_all_sums[i].map { |c| c * i })
  line.color(Nyaplot::Colors.qual[i])
end
plot.show
nil
```
